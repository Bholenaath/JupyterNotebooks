```
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sn
```

```
df = sn.load_dataset('iris')
df.head()
```

```
# Creating a simple scatter plot in Seaborn
%matplotlib inline
sn.set_palette("dark")
sn.relplot(data=df, x='petal_length', y='petal_width', hue='species',height=7, aspect=1.5, style='species')


```

```
plt.rcParams['figure.figsize'] = (11,7)
sn.set_palette('RdYlBu')
sn.set_style('ticks')
sn.lmplot(x='sepal_length'
          , y='sepal_width'
          , data=df
          , markers = 'o'
          # , size=5 # <-- deprecated
          , aspect=1.5
          , height=7
          # , sharey=True
          # , fit_reg=False
          )

# plt.xlim(0,None)
# plt.ylim(0,None)
# sn.desaturate()
```

```
plt.rcParams['figure.figsize'] = (11,7)
sn.set_palette('RdYlBu')  # <--- Like palette optoin of RdYlBu
sn.set_style('whitegrid')
sn.boxplot(data=df)

```

```
sn.set_theme('talk')
sn.set_style('whitegrid')

sn.violinplot(data=df)
```

```
sn.set_theme('paper')
sn.set_palette('icefire')
sn.set_style('darkgrid')

sn.swarmplot(data=df
             , alpha=0.7)
```

```
sn.set_theme('paper')
sn.set_palette('RdYlBu')
sn.set_style('darkgrid')

sn.violinplot(data=df
              , inner=None
              , saturation=0.3)

sn.swarmplot(data=df
             , alpha=0.7
             , color='k'
             , size=3
             , linewidth=1
             , edgecolor='white'
             )

plt.title('Iris Dataset', fontdict={'fontsize':20})
```

```
sn.set_palette('RdYlBu')
sn.heatmap(df.corr()
          , linewidths=3
          , linecolor='white')
plt.title('Iris Dataset Corelation', fontdict={'fontsize': 20})

plt.xticks(labels=None)
```

```
plt.figure(figsize=(11,7))
sn.set_palette('Pastel1')
sn.set_style('whitegrid')
sn.displot(df.sepal_length
           , rug=True
           , aspect=2
           , kind='kde')
```
